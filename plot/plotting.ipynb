# **Plotting Results for "Introduction to Computer Science" Project**



### **Introduction**

This notebook aims to present the results of the trained models for the "Introduction to Computer Science" project. The results include accuracies, metrics, loss, and other relevant information. The notebook also provides a comparative analysis of the results to draw conclusions about the performance of the models.

### **Importing Required Packages**

The first step in the analysis is to import all the necessary packages and modules required for plotting. These packages include matplotlib, seaborn, and pandas. The following code block imports these packages:

```python
import pandas as pd
import seaborn as sns
import pickle
import matplotlib.pyplot as plt

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

### **Importing Data**

To import pickle files (`.pkl`) in Python, we can use the `pickle` module. The `pickle` module implements binary protocols for serializing and de-serializing a Python object structure. It allows us to convert Python objects into a stream of bytes and vice versa. To import a pickle file, we can use the `pickle.load()` method. This method reads a pickled object representation from a file and returns the reconstituted object hierarchy specified therein.

```python
alex_path = "train_alexnet.pkl"
alex = pickle.load(open(alex_path, "rb"))

vgg_path = "train_vgg16.pkl"
vgg = pickle.load(open(vgg_path, "rb"))

resnet34_path = "train_resnet34.pkl"
resnet34 = pickle.load(open(resnet34_path, "rb"))

resnet50_path = "train_resnet50.pkl"
resnet50 = pickle.load(open(resnet50_path, "rb"))

resnet101_path = "train_resnet101.pkl"
resnet101 = pickle.load(open(resnet101_path, "rb"))
```

### **Organizing Data**

Before plotting the data, it is important to organize it in a suitable format. In this notebook, the data is organized with `EasyDict`. We then convert the data into a pandas `DataFrame` for easier plotting. Using nested list comprehensions, we can efficiently convert the data into a pandas dataframe. The following code block demonstrates how to convert the data into a pandas dataframe:




```python
alex_keys = [alex.Accuracy, alex.Precision, alex.Recall, alex.Loss_train, alex.Loss_val]
alex_list = [[tensor.item() if type(tensor) != float else tensor for tensor in key] for key in alex_keys]

vgg_keys = [vgg.Accuracy, vgg.Precision, vgg.Recall, vgg.Loss_train, vgg.Loss_val]
vgg_list = [[tensor.item() if type(tensor) != float else tensor for tensor in key] for key in vgg_keys]

resnet34_keys = [resnet34.Accuracy, resnet34.Precision, resnet34.Recall, resnet34.Loss_train, resnet34.Loss_val]
resnet34_list = [[tensor.item() if type(tensor) != float else tensor for tensor in key] for key in resnet34_keys]

resnet50_keys = [resnet50.Accuracy, resnet50.Precision, resnet50.Recall, resnet50.Loss_train, resnet50.Loss_val]
resnet50_list = [[tensor.item() if type(tensor) != float else tensor for tensor in key] for key in resnet50_keys]

resnet101_keys = [resnet101.Accuracy, resnet101.Precision, resnet101.Recall, resnet101.Loss_train, resnet101.Loss_val]
resnet101_list = [[tensor.item() if type(tensor) != float else tensor for tensor in key] for key in resnet101_keys]
```

```python
alex_df = pd.DataFrame(alex_list, index=["Accuracy", "Precision", "Recall", "Loss_train", "Loss_val"], columns=range(1,21)).T

vgg_df = pd.DataFrame(vgg_list, index=["Accuracy", "Precision", "Recall", "Loss_train", "Loss_val"], columns=range(1,21)).T

resnet34_df = pd.DataFrame(resnet34_list, index=["Accuracy", "Precision", "Recall", "Loss_train", "Loss_val"], columns=range(1,21)).T

resnet50_df = pd.DataFrame(resnet50_list, index=["Accuracy", "Precision", "Recall", "Loss_train", "Loss_val"], columns=range(1,21)).T

resnet101_df = pd.DataFrame(resnet101_list, index=["Accuracy", "Precision", "Recall", "Loss_train", "Loss_val"], columns=range(1,21)).T
```

### **Ploting Data**

Now that the data is organized, we can plot it using the `matplotlib` and `seaborn` packages. The following code blocks demonstrates how to plot the data:

```python
# Plotting accuracy curves for all models

accu_all = pd.concat([alex_df.Accuracy, vgg_df.Accuracy, resnet34_df.Accuracy, resnet50_df.Accuracy, resnet101_df.Accuracy], axis=1)
accu_all.columns = ["AlexNet", "VGG16", "ResNet34", "ResNet50", "ResNet101"]


sns.set_style("darkgrid")
sns.lineplot(data=accu_all, palette="tab10", linewidth=2, dashes=False)
plt.title("Accuracy")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Accuracy")
```

```python
# Plotting precision curves for all models

prec_all = pd.concat([alex_df.Precision, vgg_df.Precision, resnet34_df.Precision, resnet50_df.Precision, resnet101_df.Precision], axis=1)
prec_all.columns = ["AlexNet", "VGG16", "ResNet34", "ResNet50", "ResNet101"]

sns.set_style("darkgrid")
sns.lineplot(data=prec_all, palette="tab10", linewidth=2, dashes=False)
plt.title("Precision")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Precision")
```

```python
# Plotting recall curves for all models

rec_all = pd.concat([alex_df.Recall, vgg_df.Recall, resnet34_df.Recall, resnet50_df.Recall, resnet101_df.Recall], axis=1)
rec_all.columns = ["AlexNet", "VGG16", "ResNet34", "ResNet50", "ResNet101"]

sns.set_style("darkgrid")
sns.lineplot(data=rec_all, palette="tab10", linewidth=2, dashes=False)
plt.title("Recall")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Recall")
```

```python
sns.set_style("darkgrid")
sns.lineplot(data=alex_df[['Loss_train', 'Loss_val']], palette="tab10", linewidth=2)
plt.title("Training and Validation Loss of Alexnet")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Loss")
```

```python
sns.set_style("darkgrid")
sns.lineplot(data=vgg_df[['Loss_train', 'Loss_val']], palette="tab10", linewidth=2)
plt.title("Training and Validation Loss of VGG16")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Loss")
```

```python
sns.set_style("darkgrid")

sns.lineplot(data=resnet34_df[["Loss_train", "Loss_val"]], palette=["red", "red"], linewidth=2)
sns.lineplot(data=resnet50_df[["Loss_train", "Loss_val"]], palette=["orange", "orange"], linewidth=2)
sns.lineplot(data=resnet101_df[["Loss_train", "Loss_val"]], palette=["green", "green"], linewidth=2)

plt.title("Training and Validation Loss of Resnet Models")
plt.xlabel("Epoch")
plt.xticks(range(1, 21), range(1, 21))
plt.ylabel("Loss")
plt.legend(loc='upper right')
```
